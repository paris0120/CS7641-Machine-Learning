## ID3

Loop:
1. A <- Best attribute
1. Assign A as decision attribute for node
1. for each value of A create a descendent of node
1. sort training examples to the leaves
1. if examples perfectly classified
    1. stop
1. else iterate over leaves



## Bias of ID3

Two types of bias with algorithems
1. Restriction bias: hypothesis set cared about (only decision trees)
1. Preference bias: what source of hypotheses from this hyothesis set we prefer

### Inductive bias (preference bias)
ID3: prefers 
1. a tree with a good split at top
1. correct over incorrect (ones that model the data better)
    1. trees don't have good splits at the top but give you the correct answer
    1. over trees give good splits at the top but produces the wrong answer
1. prefer shorter trees to longer trees
