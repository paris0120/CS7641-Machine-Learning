## Simulated Annealing
1. Don't always improve (exploit) - sometimes you need to search (explore)
1. Metropolis-Hastings (metallurgy)
    1. repeated heating and cooling strengthens the blade
1. Algorithm
    1. For a finite set of iterations
        1. sample new point $x_t$ in N(x)
        1. Jump to new sample with probability given by an acceptance probability function $P(x, x_t, T)$
            1. $P(x, x_t, T)$ = 
                1. 1, if $f(x_t) \le f(x)$
                1. $e^{f(x_t)-f(t)\over T}$ otherwise
            1. T: tempreature
                1. Big T -> likely to accept, move freely
                1. Small T -> unlikely to accept, only uphill
1. Properties
    1. $T->0$: like hill climbing
    1. $T->\infty$: random walk
    1. decrease T slowly
        1. when T is high, wander through big valleys
        1. cooler -> valleys become boundaries, and it starts to break things up into different basins of attraction. Still able to walk small gullies
        1. give it a change to wonder to where the high value points are before cooling it so much that it can't bridge the local optimals 
    1. Pr(ending at X) = $e^{f(x)/t}\over Z_T$ Boltzmann distribution

### MIMIC: A probability model

1. try to estimate P(X)
1. start from min (uniform）
1. inprove with until get the optimal points

#### Pseudo Code
1. Generate samples from P(X): Population
1. Set $\theta_{t+1}$ to nth percentile: Fittest
1. retain only those samples s.t. $f(x) \ge \theta_{t+1}$
1. Estimate $P(X)^{\theta_{t+1}}$: Structure (how we represent these probability distributions)
1. repeat


#### Two conditions
1. We can do the estimate $P(X)^{\theta_{t+1}}$ with finite set of data
1. The probability distribution of the $\theta$ and $\theta + \epsilon$ (a slightly higher value of $\theta$) should be close

### Estimating Distributions
1. $P(x) = P(x_1|x_2,..,x_n)P(x_2|x_3,..,x_n)..P(x_n)$
1. with dependency trees
    1. $\hat P_\pi (x) = \prod P(x_i|\pi(x_i))$
    1. quadratic in the number of features 
1. use relationship to mimic covariance
1. distribution doesn't depend upon dependency trees. It's jsut the easiest way to represent distributions and capture relationships
1. allows to capture teh same kind of relationships that we get from crossover 
1. sampling is easy

### Finding dependency trees
1. $D_{KL}P||\hat P_\pi = \sum P[lg P- lg\hat p_\pi]$
    1. if same 0
    1. minimize to get p close to p hat
1. $= -h(p) + \sum h(x_i|\pi(x_i))$ 
    1. h(p) entropy 
1. minimize $J_\pi = \sum h(x_i|\pi(x_i))$
    1. minimize $-\sum h(x_i) + \sum h(x_i|\pi(x_i))$
    1. $= - \sum I(x_i;\pi(x_i))$
    1. maximize $\sum I(x_i;\pi(x_i))$ mutual information
    
    
1. fully connected graph
    1. node: x
    1. link: mutual information between them
1. maximum spinning tree 
    1. Prim and Kruskal
        1. polynomial
1. find the best distribution and dependency tree, independent of the true underlying distribution
