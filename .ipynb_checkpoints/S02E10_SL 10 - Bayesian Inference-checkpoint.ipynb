### Representing and reasoning with probabilities
1. Bayesian Network: representing and manipulating probabilistic quantities over complex spaces.


1. Joint Distribution


### Conditional independence
1. definition: X is conditional indenpendent of Y given Z
    1. if the probability distribution governing X is independent of the value of Y given the value of Z that is if
        1. $\forall x, y, z P(X=x|Y=y, Z=z) = P(X=x|Z=z)$
        1. $\forall x, y, z P(X|Y, Z) = P(X|Z)$ compact version

### Belief Network
1. AKA
    1. Bayes Nets
    1. Baysian Networks
    1. Graphical Models
1. represent conditional independence relationship between all the variables in the joint distribution graphically
    1. Nodes: variables
    1. Edges: dependencies need to be explicitly represented
        1. dependency
        1. conditional independency if not connected
    1. fill in the prior probability
    1. grow exponentially only with number of parents
1. all combinations of what the inputs are: grows exponentially with more variables (indegree)



### Sampling from the joint distribution
1. we can sample from A & B
1. To sample C we need value from sampled A & B
1. Order is: topological sort
    1. from no parents to parents
    1. can't be cyclical
    1. must be acylic
1. the joint probability for some assignement to the variables is equal to the product of all the individual values

### Inferencing Rules
1. Marginalization: represent the probability of a value by summing over some other variable and looking at the joint probabilties of those
    1. $P(x)= \sum P(x,y)$
1. Chain rule
    1. $P(x,y) = P(x)P(y|x)$
1. Bayes rule
    1. $P(y|x) = {P(x|y)P(y)\over P(x)}$

### Naive Bayes: Special Case
1. Assume that all the attributes are conditionally independent
1. $P(V|a_1, a_2, .., a_n) = \prod P(a_i|V) * P(V)/Z$
1. MAP class = argmax $P(V)\prod P(a_i|V)$


### Why Naive Bayes is cool
1. Inference is cheap
1. Few parameters 
    1. not exponential 
    1. linear
1. Estimate parameters with labeled data
    1. $P(a_i|V) = {\#a_i|v\over \#v}$ in the sample
1. Connects inference and classification
1. Empirically successful
1. No free lunch
    1. doesn't model interrelationships between attributes (ordering presend)
    1. one unseen attribute (0) spoils the whole burden g (smooth) => inductive bias
