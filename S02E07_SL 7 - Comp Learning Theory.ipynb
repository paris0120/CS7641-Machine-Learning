### PAC (Probably Approximately Correct) Learning - Error of h
1. Training error: fraction of training examples misclassified by h
1. True error: fraction of examples that would be misclassified on sample drawn from D
    1. $error_D(h) = Pr_{x~D}[c(x)\neq h(x)]$

1. C: concept class, from which the concpet that we're trying to learn comes from
1. L: Learner
1. H: Hypothesis space, the set of mappings that the learner is going to consider
1. N: $|H|$, size of hypothesis space
1. D: distribution over inputs
1. $\epsilon$: error goal $0 \leq \epsilon \leq 1/2$, the error we would like in the hypothesis no bigger than  $\epsilon$
1. $\delta$: certainty goal $0 \leq \delta \leq 1/2$, if can't meet error goal, this can provide a certainty/uncertainty goal that with probability one minus delta, the algorithm has to work (to produce a true error less than or equal to $\epsilon$).
1. $\epsilon$ and $\delta$ can't be 0, because 
    1. we sample from D
    1. we can sample high $\epsilon$ only though very rare
    1. so we have $\delta$
1. Probably Approximately Correct
    1. Probably: $1-\delta$
    1. Approximately: $\epsilon$
    1. Correct $h(x) = c(x)$
1. C is PAC-learnable by L using H 
    1. iff(if and only if) learner L will, 
    1. with probability  $1-\delta$, 
    1. output a hypothesis $h\in H$ such that 
    1. $error_D(h)\leq \epsilon$ 
    1. in time and samples polynomial in $1/\epsilon, 1/\delta$, and n
    1. it's PAC learnable if you can learn to get low error, at least with some high confidence, in time that's sort of polynomial in all the parameters

### Haussler Theorem- Bound True Error
1. bounding the true error as a function of the number of training examples 
1. Let $error_D(h_1, ... , h_k\in H)>\epsilon$: High true error |H|   
    1. How much dat do we need to "knock out" these hypotheses?
    1. $Pr_{x~D}(h_i(x)=c(x))\leq 1-\epsilon$: low probability of match
    1. given any sample set, we want to know the probability that even we draw m samples, 
        1. $h_i$ reamins consistent with c: $Pr(h_i$ consistent with c on m examples$)\leq (1-\epsilon)^m$
        1. probability of this happens (if independent iid)
    1. $Pr($at least one of $h_1,...,h_k$ consistent with c on m example$)\leq K(1-\epsilon)^m\leq |H|(1-\epsilon)^m\leq |H|e^{-\epsilon m} \leq \delta$ 
        1. $-\epsilon >= ln(1-\epsilon)$ => $(1-\epsilon)^m \leq e^{-\epsilon m}$
        1. Upper bound that version space not $\epsilon$-exhausted after m samples
        1. $ln|H|-\epsilon m \leq ln\delta, m>=1/\epsilon(ln|H|+ ln(1/\delta))$: polynomial in one over epsilon, 1/delta and size of H
1. The bound is agnostic to the distribution of sample
1. if the target concept is not in the H, we have a learning scenario referred to in the literature as agnostic, need to find one that fits nearly the best of all in H. Doesn't have to actually match the true concept, has to get close to the true in the target space
1. if |H| is infinite: hosedness
