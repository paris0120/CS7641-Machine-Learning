### Kernel Trick
1. if we push on this notion of similarity, 
    1. we care about maximizing some function that depends highly upon how different data points are alike and different
    1. can simply compute the dot product
    1. in the formulation of quadratic program in terms of W, each time in the code you want to compute $x_i^Tx_j$, if you just square it right befroe you actually used it, it would be as if you projected it into a third dimension and found a plane
    1. maximizing some function that depends highly upon how different data points are alike
    1. never need $\phi$
    
    

### Kernel
1. $W(a)=\sum a - 1/2 \sum a_ia_jy_iy_jK(x_i,x_j)$
1. $K(x_i,x_j)$ transfer xi xj to some number -> similarity
    1. inject domain knowledge into the svm learning algorithm
1. neatness
    1. you can create these kernals which can have arbitrary relationships to one another, so you can project into some higher dimensional space, in which your points are linearly separable
    1. because you use the kernel function to represent your domain knowledge, you don't have to do the computation of transformating the points into this higher dimensional space
1. polynomial kernal to represent polynomial functions: $K=(X^TY+c)^P$
1. radial basis kernal: $K = e^{-||x-y||^2/2\sigma^2}$
1. sigmoid: $K=tanh(ax^Ty+\theta)$
