### Boosting
1. focuse on hardest examples
1. weighted mean
1. Error: $Pr_D[h(x) \neq c(x)]$
    1. Pr: probablity
    1. D: distribution
    1. h: hypothesis.
    1. c: true underlying concept
    1. given the probabilty and underlined distribution, that I will disagree with th true concept on some particular instance X 
1. Weak learner
    1. no matter what the distribution is over your data, will do better than chance when it tries to learn labels on that data (error rate less than half) 
    1. $ \forall_D Pr_D[*]<=1/2-\epsilon$

1. red x is not correct label for the x
1. green check for right label
1. come up with the distribution over the 4 different examples such that a learning algorithm that has to choose between one of those 3 hypotheses wil in fact be able to find one  better than chance 


1. Given training {X, Y}, Y in {-1, 1}
1. For t = 1 + T
    1. construct $D_t$
    1. find weak classifier $h_t(x)$ with small error $\epsilon_t = Pr_D[h_t(x)\neq y_t]$
1. output $H_{final}$

1. $D_1(i) = 1/n$
1. $D_{t+1}(i) = {{D_t(i)e^{-\alpha_t y_i h_t(x_i)}}\over Z_t}$
    1. where $\alpha_t = {1 \over 2} ln{{1-\epsilon_t}\over \epsilon_t}$: now good a particular weak hypothesis was
    1. D: distribution/probability
    1. $h_t$: returns either -1 or 1
    1. $y_i$: label, either -1 or 1
    1. $a_t$: constant postive number
    1. $\epsilon$: between 0-1 
    1. Z: normalization constant
    1. if $h_t$ agrees $y_t$, how D changes
        1. depends
            1. if agree and disagree at least some of them, at least one other example, D decreases
            1. otherwise unchanged
    1. if ht and yt disagree => increase: put more weights when geting rong
    1. the ones getting wrong are getting more weight
    1. we are guaranteed or at least we've assumed that we have a weak learner that will always do better than chance (get soem one that were wrong right)
1. $H_{final}(x) = sgn(\sum \alpha_t h_t (x))$
